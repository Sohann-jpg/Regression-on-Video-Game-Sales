#**About The Dataset**

The dataset I have chosen for Regression task contains a list of video games with sales greater than 100,000 copies worldwide. There are 16597 games listed in the dataset and they are ranked according to the number of total sales they have worldwide. 

This dataset contains 11 columns and they are listed below: 

1.   "Rank" - Ranking of games based on their sales.
2.   "Name" - Contains the name of the games.
3.   "Platform" - Platform of the games release (i.e. PC,PS4, etc.)
4.   "Year" - Date of the game release.
5.   "Genre" - Genre of the game (For eg: Racing, Sports, RPG, etc.)
6.   "Publisher" - Name of the pubishing company or the publisher of the game. 
7.   "NA_Sales" - Sales in North America.
8.   "EU_Sales" - Sales in Europe.
9.   "JP_Sales" - Sales in Japan.
10.  "Other_Sales" - Sales in the rest of the world.
11.  "Global_Sales" - Sales all of the world (Total Sales)


*****All sales are in Millions.*****

Predicting some trends based off of the data. We can carry out linear regression to get an idea of how global sales figures could end up based on North American sales figures. Firstly, separating data into train and test sets and setting North American sales as our X variable and global sales as our Y variable, and then doing train/test split.



---



```python
## Carrying out linear regression and predicting the global sales of games
## based off of the sales in North America

X = df.iloc[:, 6].values
y = df.iloc[:, 10].values

## train test split and spliting the dataframe

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8)
```

The data needs to be reshaped in order to be compatible with Linear Regression, so reshaping them into two long 2D arrays that have as many rows as necessary and a single column and applying the data in the Linear Regression function.



---




```python
## Reshaping the data into long 2D arrays with 1 column and as many rows as necessary
X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

Checking to see how our regression algorithm performed. Plotting the correlation between the variableâ€™s training data, and plotting the line of best fit from our regressor. Then doing the same thing for our testing data. Essentially weâ€™re looking to see how the regression line fits both the training and testing data.



```python
def plot_regression(classifier):

    plt.scatter(X_train, y_train,color='blue')
    plt.plot(X_train, classifier.predict(X_train), color='red')
    plt.title('(Training set)')
    plt.xlabel('North America Sales')
    plt.ylabel('Global Sales')
    plt.show()

    plt.scatter(X_test, y_test,color='blue')
    plt.plot(X_train, classifier.predict(X_train), color='red')
    plt.title('(Testing set)')
    plt.xlabel('North America Sales')
    plt.ylabel('Global Sales')
    plt.show()
    
plot_regression(lin_reg)
print("Training set score: {:.2f}".format(lin_reg.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lin_reg.score(X_test, y_test)))
```

#Findings
The regression lines should look approximately the same, and indeed they look fairly similar. The training set regression shows approximately 70 million sales for 40 million North American sales, while the test set regression may be just a little higher. Weâ€™ll also print the scores on the training and test sets, and see that our Linear Regression implementation had similar, though slightly worse accuracy on the testing set.
